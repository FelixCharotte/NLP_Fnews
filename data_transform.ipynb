```python
import feedparser

def get_news_pages(stock_symbol, num_pages):
    rss_url = f"https://finance.yahoo.com/rss/headline?s={stock_symbol}"
    all_news = []
    
    for page in range(1, num_pages + 1):
        page_url = f"{rss_url}&page={page}"
        feed = feedparser.parse(page_url)
        
        for entry in feed.entries:
            news_item = {
                'title': entry.title,
                'link': entry.link,
                'published': entry.published
            }
            all_news.append(news_item)
    
    return all_news

# Example usage:
stock_symbol = "AAPL"  # Change this to the stock symbol you're interested in
num_pages = 150  # Number of pages to fetch (adjust as needed)
all_news = get_news_pages(stock_symbol, num_pages)
for news_item in all_news:
    print("Title:", news_item['title'])
    print("Link:", news_item['link'])
    print("Published:", news_item['published'])
    print("---------------------")
```

```python
from datetime import datetime

def convert_date(date_str):
    # Conversion de la chaîne de date en objet datetime
    date_object = datetime.strptime(date_str, '%a, %d %b %Y %H:%M:%S %z')
    
    # Formater la date dans un format simple
    simple_date = date_object.strftime('%Y %H:%M:%S')
    
    return simple_date
```

```python
import yfinance as yf

def get_historical_prices(stock_symbol, start_date, end_date):
    stock_data = yf.download(stock_symbol, start=start_date, end=end_date, interval='1m')
    return stock_data

# Exemple d'utilisation :
stock_symbol = "AAPL"  # Symbole boursier d'Apple
start_date = "2024-04-18"  # Date de début de l'historique des prix
end_date = "2024-04-24"  # Date de fin de l'historique des prix

historical_prices = get_historical_prices(stock_symbol, start_date, end_date)
print(historical_prices)

```
